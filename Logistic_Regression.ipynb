Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.

 - Simple Linear Regression (SLR) is a statistical technique used to model the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line to the observed data.

**Purpose**:

To understand how changes in X affect Y

To predict the value of Y for a given value of X

To quantify the strength and direction of the relationship between variables

Question 2: What are the key assumptions of Simple Linear Regression?

 - The main assumptions are:

**Linearity** ‚Äì The relationship between X and Y is linear

**Independence** ‚Äì Observations are independent of each other

**Homoscedasticity** ‚Äì Constant variance of errors across all values of X

**Normality of errors** ‚Äì Residuals are normally distributed

**No multicollinearity** ‚Äì Only one predictor (automatically satisfied in SLR)

Question 7: Name and briefly describe three common evaluation metrics for regression models.

 - **Mean Absolute Error (MAE):**
Average of absolute differences between actual and predicted values

 - **Mean Squared Error (MSE):**
Average of squared prediction errors (penalizes large errors more)

 - **Root Mean Squared Error (RMSE):**
Square root of MSE; interpretable in the same unit as Y

Question 8: What is the purpose of the R-squared metric in regression analysis?

 - R-squared (R¬≤) measures how much of the variance in the dependent variable is explained by the model.

Value ranges from 0 to 1

Higher value ‚Üí better model fit

Example: R¬≤ = 0.80 means 80% of the variation in Y is explained by X

Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

# Create and fit model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
